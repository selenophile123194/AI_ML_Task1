# Task
Load the Iris dataset, inspect its structure and summary statistics, and visualize the relationships between features, value distributions, and potential outliers using scatter plots, histograms, and box plots.

## Load the dataset

### Subtask:
Load the Iris dataset into a pandas DataFrame.


**Reasoning**:
The subtask requires loading the Iris dataset into a pandas DataFrame, which involves importing necessary libraries, loading the dataset, creating a DataFrame, and assigning column names.



```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a pandas DataFrame
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display the first few rows of the DataFrame to verify
print(df_iris.head())
```

## Inspect the dataset

### Subtask:
Print the shape, column names, and the first few rows, and use `.info()` and `.describe()` for summary statistics.


**Reasoning**:
Print the shape, column names, and first few rows, and use `.info()` and `.describe()` to inspect the DataFrame.



```python
print("Shape of the DataFrame:", df_iris.shape)
print("\nColumn names:", df_iris.columns)
print("\nFirst 5 rows:\n", df_iris.head())
print("\nDataFrame Info:")
df_iris.info()
print("\nDescriptive statistics:")
df_iris.describe()
```

## Visualize the dataset

### Subtask:
Create scatter plots, histograms, and box plots to visualize the data trends and distributions.


**Reasoning**:
Generate scatter plots for all pairs of features using seaborn's pairplot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df_iris)
plt.show()
```

**Reasoning**:
Generate histograms for each feature to visualize their distributions.



```python
df_iris.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

**Reasoning**:
Generate box plots for each feature to visualize their distributions and potential outliers.



```python
df_iris.plot(kind='box', subplots=True, layout=(2, 2), figsize=(10, 8), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```
